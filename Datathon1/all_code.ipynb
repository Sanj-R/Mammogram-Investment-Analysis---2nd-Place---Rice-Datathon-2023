```python
# This function computes the best-fit line of a set of points in R^2, returning its slope and y-intercept.
# The mathematical formula for best-fit line was obtained online.

def best_fit_line(xs, ys):
    N = len(xs)
    sum_x = 0
    sum_y = 0
    sum_x2 = 0
    sum_y2 = 0
    sum_xy = 0
    for i in range(0, len(xs)):
        x = xs[i]
        y = ys[i]
        sum_x += x
        sum_y += y
        sum_x2 += x*x
        sum_y2 += y*y
        sum_xy += x*y
    m = (N * sum_xy - sum_x*sum_y)/(N * sum_x2 - (sum_x*sum_x))
    b = (sum_y - m * sum_x) / N
    return (m,b)
```

```python
# Given list of population from dataset years 2010 (Apr.), 2010 (July), 2011 (July), ..., 2019 (July), extrapolate the 2023 (Jan.) population via linear regression.

def est_pop_2023_asg(ys):
    (m,b) = best_fit_line([0, 0.25, 1.25, 2.25, 3.25, 4.25, 5.25, 6.25, 7.25, 8.25, 9.25], ys)
    return m * 12.75 + b
```

```python
# Compute the expected 2023 populations segregated by age, state, and sex.

male_state_age_year_bfe = {}
female_state_age_year_bfe = {}
for state in male_state_age_year:
    male_state_age_year_bfe[state] = {}
    for age in male_state_age_year[state]:
        male_state_age_year_bfe[state][age] = est_pop_2023_asg(male_state_age_year[state][age])
for state in female_state_age_year:
    female_state_age_year_bfe[state] = {}
    for age in female_state_age_year[state]:
        female_state_age_year_bfe[state][age] = est_pop_2023_asg(female_state_age_year[state][age])
print(male_state_age_year_bfe['Alabama']) # NOTE ADDED ['Alabama'] POST-COMPETITION TO AVOID LONG OUTPUT FOR GITHUB.
print(female_state_age_year_bfe['Alabama']) # NOTE ADDED ['Alabama'] POST-COMPETITION TO AVOID LONG OUTPUT FOR GITHUB.
```

```python
# We find research data on new breast cancer case rates for each sex, for each 5-years-long age group
# (from https://www.cancerresearchuk.org/health-professional/cancer-statistics/statistics-by-cancer-type/breast-cancer/incidence-invasive#heading-One)
# and store these in fr and mr. We then assign the same value to each year within each 5-years-long age group.

# research data on male, female breast cancer rates.
fr = [0,0,0,.1,1.6,11.5,31.2,65.8,124.6,214.8,279.8,285.5,337.9,412.3,372.7,403,430.4,447.7]
mr = [0,0,0,0,0,.1,0,.1,.4,.5,1.2,1.3,2.1,3.6,3.9,5,6.1,8.8]

# assign the same value to each year within each 5-years-long age group.
female_rates_by_age = {}
male_rates_by_age = {}
for i in range(0, 86, 5):
    female_rates_by_age[i] = fr[i//5]
    female_rates_by_age[i+1] = fr[i//5]
    female_rates_by_age[i+2] = fr[i//5]
    female_rates_by_age[i+3] = fr[i//5]
    female_rates_by_age[i+4] = fr[i//5]
    male_rates_by_age[i] = mr[i//5]
    male_rates_by_age[i+1] = mr[i//5]
    male_rates_by_age[i+2] = mr[i//5]
    male_rates_by_age[i+3] = mr[i//5]
    male_rates_by_age[i+4] = mr[i//5]
print(female_rates_by_age)
print(male_rates_by_age)
```

```python
# Using the male and female cancer rate data, we compute a cancer risk metric for each sex in each state, which is
# the predicted number of each age group of that sex in the state in 2023 times the cancer risk rate for that sex, age group.

state_cancer_risk_male = {}
for state in male_state_age_year_bfe:
    state_cancer_risk_male[state] = 0
    for age in male_state_age_year_bfe[state]:
        state_cancer_risk_male[state] += male_state_age_year_bfe[state][age] * male_rates_by_age[age]
state_cancer_risk_female = {}
for state in female_state_age_year_bfe:
    state_cancer_risk_female[state] = 0
    for age in female_state_age_year_bfe[state]:
        state_cancer_risk_female[state] += female_state_age_year_bfe[state][age] * female_rates_by_age[age]
print(state_cancer_risk_male)
print(state_cancer_risk_female)
```

```python
# We add together the sex-specific cancer risk metrics for each state to get the overall cancer risk metric for each state.

state_cancer_risk = {}
for state in state_cancer_risk_male:
    state_cancer_risk[state] = state_cancer_risk_male[state] + state_cancer_risk_female[state]
print(state_cancer_risk)
```

```python
# We sort the state risk metrics in non-increasing order.
# States with higher risk metrics are proportionally more likely to have more new cancer cases.

sorted_states_by_risk = []
for state, risk in state_cancer_risk.items():
    sorted_states_by_risk.append((state, risk))
sorted_states_by_risk.sort(key= (lambda x: x[1]))
sorted_states_by_risk.reverse()
for staterisk in sorted_states_by_risk:
    print(staterisk[0], staterisk[1])
```
